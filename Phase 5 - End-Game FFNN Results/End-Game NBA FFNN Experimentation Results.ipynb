Analysis of Features Not Appearing in SHAP Analysis
The features BLK, PF, FG3_PCT, FT_PCT, FG_PCT, AST, and HOME do not appear in the top 15 SHAP-ranked features across any of the 10 models. This absence provides important insights about their limited influence in determining NBA game outcomes within the scope of your "Nothing But Net" project.

Feature-Specific Observations
BLK (Blocks):

Expected Impact: Blocks are often considered defensive metrics but are less frequent events compared to steals (STL).
Reason for Exclusion: SHAP rankings emphasize STL and Defensive_Index, suggesting that steals may capture more relevant defensive contributions to game outcomes. Blocks might have a lower correlation with results or be too infrequent to influence predictions significantly.
PF (Personal Fouls):

Expected Impact: Personal fouls can disrupt the flow of the game and lead to free throw opportunities for opponents.
Reason for Exclusion: Metrics like FTA (Free Throws Attempted) and EFGPCT (Effective Field Goal Percentage) likely capture the downstream effects of fouls more directly. The fouls themselves may not provide as much predictive power.
FG3_PCT (Three-Point Field Goal Percentage):

Expected Impact: Three-point efficiency is critical in modern basketball.
Reason for Exclusion: SHAP ranks emphasize volume metrics like FG3A (Three-Point Attempts) instead. This suggests that total attempts may correlate more strongly with game outcomes than efficiency alone.
FT_PCT (Free Throw Percentage):

Expected Impact: High free throw percentage can influence close games.
Reason for Exclusion: Volume metrics like FTA (Free Throws Attempted) and PTS (Points) appear in the SHAP rankings, indicating that the sheer number of opportunities might be more predictive than efficiency.
FG_PCT (Field Goal Percentage):

Expected Impact: General shooting efficiency is a common metric for evaluating team performance.
Reason for Exclusion: Similar to FG3_PCT, volume metrics like FGA (Field Goals Attempted) and EFGPCT appear in the rankings instead. These likely capture a broader picture of scoring opportunities.
AST (Assists):

Expected Impact: Assists are often used as a proxy for team cohesion and offensive ball movement.
Reason for Exclusion: SHAP rankings include related metrics like TO_to_AST_Ratio (Turnover-to-Assist Ratio), which incorporates assists in the context of turnovers. This suggests that the combination of assists and turnovers is more predictive than assists alone.
HOME (Home Advantage):

Expected Impact: Home-court advantage is widely recognized as influential in sports outcomes.
Reason for Exclusion: The SHAP analysis may implicitly capture home advantage through metrics like POSSESSIONS, PACE, or PPM (Points Per Minute), which can vary significantly depending on the location. Alternatively, its exclusion might indicate a weak correlation between home games and game results in your dataset.

Key Takeaways

Importance of Volume vs. Efficiency Metrics:

Features like FG3_PCT, FT_PCT, and FG_PCT emphasize efficiency, but SHAP prioritizes volume metrics like FGA and FG3A. This indicates that the number of attempts is a stronger predictor than efficiency percentages in your dataset.
Defensive Metrics Preference:

While STL and Defensive_Index dominate SHAP rankings, BLK is excluded. This suggests that other defensive actions are more closely tied to game outcomes.
Composite Metrics Matter:

Metrics like TO_to_AST_Ratio and EFGPCT appear instead of standalone metrics like AST or PF, highlighting the importance of combined or contextualized statistics.
