```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import datetime
```

```python
nba_data = pd.read_csv("finalised_NBA_Data.csv")
```

```python
nba_data.drop(["TEAM_ABBREVIATION","TEAM_NAME","MATCHUP","HOME_TEAM","AWAY_TEAM"],axis=1,inplace=True)
```

```python
nba_data["GAME_DATE"] = pd.to_datetime(nba_data["GAME_DATE"])
```

```python
# Extract year from GAME_DATE
nba_data['Year'] = nba_data['GAME_DATE'].dt.year

# Group by Year and WL, then calculate the mean of PTS
yearly_grouped_PTS_data = nba_data.groupby(['Year', 'WL'])['PTS'].mean().reset_index()

# Pivot the data to have separate columns for WL = 'W' and WL = 'L'
yearly_pivoted_data = yearly_grouped_PTS_data.pivot(index='Year', columns='WL', values='PTS')

# Plot the yearly mean PTS
plt.figure(figsize=(12, 6))
plt.plot(yearly_pivoted_data.index, yearly_pivoted_data['W'], label='Win (WL=W)', linewidth=2 , marker='x')
plt.plot(yearly_pivoted_data.index, yearly_pivoted_data['L'], label='Loss (WL=L)', linewidth=2, linestyle='--',marker='x')

# Add plot labels and title
plt.title('Yearly Mean PTS (Wins vs. Losses)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Mean Points (PTS)', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Ensure the x-axis ticks display every year
plt.xticks(ticks=yearly_pivoted_data.index, labels=yearly_pivoted_data.index, rotation=45)


# Display the plot
plt.show()
```

```python
# Extract year from GAME_DATE 
nba_data['Year'] = nba_data['GAME_DATE'].dt.year

# List of metrics to plot
metrics = ['BLK','STL']

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, figsize=(14, 12))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Loop through each metric and plot
for i, metric in enumerate(metrics):
    # Group by Year and WL, then calculate the mean of the current metric
    yearly_grouped_data = nba_data.groupby(['Year', 'WL'])[metric].mean().reset_index()

    # Pivot the data to have separate columns for WL = 'W' and WL = 'L'
    yearly_pivoted_data = yearly_grouped_data.pivot(index='Year', columns='WL', values=metric)

    # Plot on the respective subplot
    axes[i].plot(yearly_pivoted_data.index, yearly_pivoted_data['W'], label='Win (WL=W)', linewidth=2, marker='x')
    axes[i].plot(yearly_pivoted_data.index, yearly_pivoted_data['L'], label='Loss (WL=L)', linewidth=2, linestyle='--', marker='x')

    # Add titles, labels, and legend
    axes[i].set_title(f'Yearly Mean {metric} (Wins vs. Losses)', fontsize=14)
    axes[i].set_xlabel('Year', fontsize=12)
    axes[i].set_ylabel(f'Mean {metric}', fontsize=12)
    axes[i].legend()
    axes[i].grid(True)

    # Set ticks to display every year
    axes[i].set_xticks(yearly_pivoted_data.index)
    axes[i].set_xticklabels(yearly_pivoted_data.index, rotation=45)

# Tight layout to avoid overlap
plt.tight_layout()
plt.show()
```
