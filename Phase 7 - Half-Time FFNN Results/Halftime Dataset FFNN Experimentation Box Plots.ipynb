```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Halftime Full - Reduced Individual Model Comparision.csv')
```

```python
data.head(15)
```

```python
# Generate box plots using the DataFrame
metrics_pairs = [
    ("Full Accuracy", "Reduced Accuracy"),
    ("Full AUC", "Reduced AUC"),
    ("Full Loss", "Reduced Loss"),
    ("Full Precision", "Reduced Precision"),
    ("Full Recall", "Reduced Recall")
]

for full_metric, reduced_metric in metrics_pairs:
    plt.figure(figsize=(6, 4))
    plt.boxplot(
        [data[full_metric], data[reduced_metric]],
        labels=[full_metric, reduced_metric],
        patch_artist=True
    )
    plt.title(f'Boxplot for {full_metric} and {reduced_metric} for Half-Time Dataset Models')
    plt.ylabel('Values')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```
